```python
import numpy as np
from PIL import Image
import cv2
import imutils
import matplotlib.pyplot as plt
%matplotlib inline

import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract'
```

```python
image_grey = cv2.imread('../images/captcha_images/samples/226md.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image_grey, cmap='gray')
```

```python
(thresh, image_bin) = cv2.threshold(image_grey, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
thresh
plt.imshow(image_bin, cmap='gray')
```

```python
# upscale image
print('Original Dimensions : ',image_bin.shape)
 
scale_percent = 500 # percent of original size
width = int(image_bin.shape[1] * scale_percent / 100)
height = int(image_bin.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
image_resized = cv2.resize(image_bin, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ', image_resized.shape)
plt.imshow(image_resized, cmap='gray')
```
